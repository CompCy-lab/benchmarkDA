### Read the data

```python
# read the data and filter the markers
adata = sc.read_h5ad('../../data/real/bcr-xl/bcr_xl.h5ad')

marker_in_use = ['CD3', 'CD45', 'pNFkB', 'pp38', 'CD4', 'CD20', 'CD33', 'pStat5', 'CD123', 'pAkt', 'pStat1', 'pSHP2',
         'pZap70', 'pStat3', 'CD14', 'pSlp76', 'pBtk', 'pPlcg2', 'pErk', 'pLat', 'IgM', 'pS6', 'HLA-DR', 'CD7']
adata.var.index = adata.var['pns_label']
adata.var.index.name = 'index'
adata = adata[:,marker_in_use]

# reset the index
adata.obs = adata.obs.reset_index(drop=True)
adata.obs.index.name = 'cell'
adata.obs.set_index('cell_' + adata.obs.index.astype(str), inplace=True)

seed = 0
co_factor = 5.
adata.X = np.arcsinh(adata.X / co_factor)
```

```python
# set sample id and prepare the data format for cna approach
def fcs2id(x):
    _id, _type = x.strip().split('_')[-2:]
    _id = int(_id[-1])
    _type = _type.strip('.fcs')
    if _type == 'BCR-XL':
        _id = _id * 2 - 1
    else:
        _id = _id * 2
    return _id

adata.obs['id'] = ""
adata.obs['id'] = adata.obs['fcs_files'].map(fcs2id).astype(int)

adata.obs['case'] = np.nan
adata.obs['case'] = adata.obs['phenotype'].map({'BCR-XL': 1, 'Ref': 0}).astype(int)
adata.obs
```
